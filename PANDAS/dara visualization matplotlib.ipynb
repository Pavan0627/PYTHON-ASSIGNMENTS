```python
CourseTaught=['Msc CS','Msc DS','Msc  CA','IMCA']
girls=[25,12,24,11]
boys=[40,22,40,20]
x=np.arange(len(CourseTaught))
width=0.35
##intialize a figure and an axes
fig,ax=plt.subplots()
bar1=ax.bar(x-width/2,boys,width,label="BOYS",color='skyblue')
bar2=ax.bar(x+width/2, girls,width,label="GIRLS",color='pink')
ax.set_xticks(x)
ax.set_xticklabels(CourseTaught)
ax.legend()
plt.show()
```

## line graph

```python
np_df=pd.read_csv('national_parks.csv')
np_df.head()

```

```python
plt.plot(np_df['GrandCanyon'])
```

```python
fig,ax1=plt.subplots(figsize=(8,8))
ax2=ax1.twinx()

ax1.plot(np_df['Year'],np_df['GrandCanyon'],lw=3,color='skyblue')
ax2.plot(np_df['Year'],np_df['BryceCanyon'],lw=4,color='orange')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number of visitors for GrandCanyon',fontsize=14)
ax1.tick_params(axis='y')

ax2.set_ylabel('Number of visitors for BryceCanyon',fontsize=14)
ax2.tick_params(axis='y')

fig.suptitle("visitors of national parks",fontsize=18)
fig.legend
```
